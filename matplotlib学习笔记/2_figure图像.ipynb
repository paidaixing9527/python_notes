```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 在[-3,3]之间平均生成50点
x = np.linspace(-3, 3, 50)

# 定义两个方程
y1 = 2 * x + 1
y2 = x ** 2
```

```python
# 定义一个figure
plt.figure(num=3, figsize=(8, 5))
plt.plot(x, y1)
plt.show()
```

```python
# 再定义一个figure
plt.figure(num=1,figsize=(10, 6))
plt.plot(x, y1)
plt.plot(x, y2,color='red', linewidth=3, linestyle='--')
plt.show()
```
