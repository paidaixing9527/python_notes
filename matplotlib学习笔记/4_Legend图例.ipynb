```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 50)

y1 = 2 * x + 1

y2 = x ** 2
```

```python
plt.figure(num=1, figsize=(10,6))

l1, = plt.plot(x, y1, label='down')

l2, = plt.plot(x, y2, label='up', color='red', linewidth=2, linestyle='--')

# 设置x轴的范围为 -1 到 2
plt.xlim((-1, 2))

# 设置y的范围为 -2 到 3
plt.ylim((-2, 3))

# 设置 x 轴的 label
plt.xlabel('I am x')

# 设置 y 轴的 label
plt.ylabel('I am y')

new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)

# 将 x 轴的 ticks 设置为 new_ticks
plt.xticks(new_ticks)

# 设值 y 轴的 ticks
plt.yticks([-2, 0, 1.8, 3],[r'$really\ bad$', r'$bad\ \alpha$', r'$good$',r'$really\ good$'])

plt.legend(handles=[l1,l2,], labels=['aaa','bbb'], loc='best')

plt.show()
```
