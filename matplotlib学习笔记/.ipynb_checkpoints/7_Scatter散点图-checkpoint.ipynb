```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1024
n = 1024

X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)

# 
T = np.arctan2(Y, X)

plt.figure(num=1, figsize=(10,10))

# s:  c alpha
plt.scatter(X, Y, s=75, c=T, alpha=0.5)

plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))

# ticks
plt.xticks(())
plt.yticks(())

plt.show()
```
