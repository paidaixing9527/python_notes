```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(num=1,figsize=(5,5))

# 将整个figure分隔成两行两列，在第一个位置显示一张小图
plt.subplot(2,2,1)
plt.plot([0,1],[0,1])
# 第二张小图
plt.subplot(2,2,2)
plt.plot([0,1],[0,2])
# 第三张小图
plt.subplot(2,2,3)
plt.plot([0,1],[0,3])
# 第四张小图
plt.subplot(2,2,4)
plt.plot([0,1],[0,4])


plt.show()
```
