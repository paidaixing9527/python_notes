```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1024个点
n = 1024

X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)

# 颜色
T = np.arctan2(Y, X)

plt.figure(num=1, figsize=(10,10))

# s: 点大小， c：点颜色， alpha：点透明度
plt.scatter(X, Y, s=75, c=T, alpha=0.5)

plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))

# 隐藏所有的ticks
plt.xticks(())
plt.yticks(())

plt.show()
```
